```python
import pandas as pd
```

```python
df = pd.read_excel("output.xlsx")
deals_df = pd.read_csv("clear_deals.csv")
```

```python
merged = df.merge(deals_df, left_on="Полный адресс", right_on="адрес_корпуса", how="inner")
res = merged.groupby("Полный адресс")[["суммарная_площадь_сделок", "реальная_сумма_бюджета"]].sum().reset_index()
res["средняя цена за кв. м, ₽"] = res["реальная_сумма_бюджета"] / res["суммарная_площадь_сделок"]
```

```python
classes = res.merge(deals_df, left_on="Полный адресс", right_on="адрес_корпуса").groupby("Полный адресс")["класс"].first().reset_index(drop=True)
res["класс_проекта"] = classes
```

```python
res = res.merge(df, on="Полный адресс")[["средняя цена за кв. м, ₽", "класс_проекта", "Долгота", "Широта"]]
res.to_csv("geodata_for_map.csv", index=False)
```
